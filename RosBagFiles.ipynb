```python
from rosbags.rosbag2 import Reader
from rosbags.serde import deserialize_cdr
```

```python
# create reader instance and open for reading
with Reader('/home/dschmoeller/ros2_ws/bag_files/rosbag2_2023_02_10-07_55_37') as reader:
    # topic and msgtype information is available on .connections list
    for connection in reader.connections:
        print('Communication information ----------------------')
        print(connection.id)
        print(connection.topic)
        print(connection.msgtype)
        print(connection.msgcount)
        print(connection.ext.serialization_format)
        print('------------------------------------------------')
        print()

    # iterate over messages
    for connection, timestamp, rawdata in reader.messages():
        if connection.topic == '/nav_vel':
            # Deserialize raw data 
            msg = deserialize_cdr(rawdata, connection.msgtype)
            print('timestamp:', timestamp)
            print(msg.linear)
            print(msg.angular)
            print()
    # messages() accepts connection filters
    connections = [x for x in reader.connections if x.topic == '/nav_vel']
    for connection, timestamp, rawdata in reader.messages(connections=connections):
        msg = deserialize_cdr(rawdata, connection.msgtype)
        #print(msg.header.frame_id)
        #print(connection)
        #print(type(msg))
```
